# Adding a template

```python
from langchain_core.prompts import ChatPromptTemplate
chat_template = ChatPromptTemplate.from_messages(
        [
            ("system","you are a helpful assistant that gives a one line defination of the word enetered by user"),
            ("human","{user_input}"),
        ]
)
messages = chat_template.format_messages(user_input='sodium')
messages
```

# step 2: Defining model

```python
from langchain_ollama import ChatOllama
llm=ChatOllama(
    model='alibayram/smollm3:latest',
    temprature=0
)
```

```python
ai_msg=llm.invoke(messages)
ai_msg
```

# Step 3: Creating Pipeline

```python
from langchain_core.output_parsers import StrOutputParser
chain = chat_template | llm | StrOutputParser()
```

# Step 4: Invoking Chain

```python
chain.invoke({"user_input":"Onomatopoeia"})
```

# making RAG

```python
from langchain_community.document_loaders import TextLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_community.embeddings import OllamaEmbeddings
from langchain_community.vectorstores import FAISS
from langchain_ollama import ChatOllama
```

# Loading the Training Document

```python
raw_documents = TextLoader("C:/Users/Lenovo/Desktop/RAG_text.txt",encoding="utf-8").load()
```

```python
raw_documents
```

# Splitting the documents in to chunks 

```python
text_splitter = RecursiveCharacterTextSplitter(chunk_size=300, chunk_overlap=20)
documents=text_splitter.split_documents(raw_documents)
```

# 5 chunks were made of above document 

```python
len(documents)
```

```python
print(documents[0])
```

# Vector Embedding this Chunks

## this local host "http://localhost:11434" is where all requests and commands  are delivered for local machine it does'nt go outside

```python
oembed = OllamaEmbeddings(base_url="http://localhost:11434",model="mxbai-embed-large")
```

## mxbai-embed-large is the embedding model used for vector embedding (costlier process)

```python
%pip install langchain-community faiss-cpu
```

## FAISS is the special storage where we store vector embedding , because we cannot store vector embeddings in normal storage

```python
db = FAISS.from_documents(documents, embedding=oembed)
```

```python
query = "What is the price and release date of the new robot?"

# 'k=1' retrieves the single most relevant document chunk
docs = db.similarity_search(query, k=1)
```

## retrived one relevant chunk according to query

# Start making RAG

```python
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser
from langchain_core.runnables import RunnablePassthrough
```

## Step 1 Making the Template

```python
template="""
You are an expert assistant. Answer the user's question based ONLY on the provided context.

Context: {context}

Question: {question}
"""
prompt = ChatPromptTemplate.from_template(template)
```

## Defing the model

```python
model = ChatOllama(
model="alibayram/smollm3:latest",
    temprature=0
)
```

## Retriving the chunks in a variable

```python
retriever=db.as_retriever()
```

```python
def format_docs(docs):
    return"\n\n".join([d.page_content for d in docs])
```

## Making a Chain 

```python
chain = (
{"context":retriever | format_docs,"question": RunnablePassthrough()}
| prompt
| model
| StrOutputParser()
)
```

## asking out of context Question

```python
chain.invoke("what is text embedding and how does langchain help in doing it")
```
